```python
import cv2
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
img = cv2.imread("./test_images/maria_sharapova_2.jpg")
img.shape
```

```python
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray.shape
```

```python
plt.imshow(gray , cmap = 'gray')
```

## Face detection using HaarCascade

```python
face_cascade = cv2.CascadeClassifier('./opencv/haarcascades/haarcascade_frontalface_default.xml')

faces = face_cascade.detectMultiScale(gray , 1.3 , 5)
faces
```

```python
(x,y,w,h) = faces[0]
x,y,w,h
```

```python
gray.shape
```

```python
roi_image = gray[y:y+h , x:x+w ]
```

```python
top_left_point = (x,y)
bottom_right_point=(x+w , y+h)
```

```python
face_image = cv2.rectangle(gray , top_left_point , bottom_right_point , (255,0,0) , 3)
plt.imshow(face_image)
```

## Eyedetection using HaarCascade

```python
eye_cascade = cv2.CascadeClassifier('./opencv/haarcascades/haarcascade_eye.xml')
eyes = eye_cascade.detectMultiScale(roi_image)
eyes
```

```python
for eye in eyes:
    (x,y,w,h) = eye
    top_left_point = (x,y)
    bottom_right_point=(x+w , y+h)
    eye_image = cv2.rectangle(roi_image , top_left_point , bottom_right_point , (0,255,0) , 2)
    plt.imshow(eye_image)
```

### Let us save the region of interest for further reference

```python
plt.imshow(roi_image)
roi_image
```

```python
def get_cropped_image_if_2_eyes(image_path):
    img=cv2.imread(image_path)
    gray = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray , 1.3 , 5)
    for(x,y,w,h) in faces :
        roi_gray = gray[y:y+h , x:x+w] 
        roi_color= img[y:y+h , x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray)
        if len(eyes) >=2 :
            return roi_color
```

```python
roi_color = get_cropped_image_if_2_eyes("./test_images/maria_sharapova_2.jpg")
plt.imshow(roi_color)
```

### Let us create a new folder cropped in which we will store all the cropped dataset

```python
path_to_data="./dataset"
path_to_cropped_dataset = "./dataset/cropped"
```

```python
import os 
img_dirs =[]
for entry in os.scandir(path_to_data):
    if entry.is_dir() :
        img_dirs.append(entry.path)

img_dirs
```

```python
import shutil
if os.path.exists(path_to_cropped_dataset):
    shutil.rmtree(path_to_cropped_dataset)
os.mkdir(path_to_cropped_dataset)
```

```python
for img_dir in img_dirs:
    count =1
    celebrity_name = img_dir.split('\\')[-1]
    for entry in os.scandir(img_dir):
        roi_color=get_cropped_image_if_2_eyes(entry.path)
        if roi_color is not None :
            cropped_folder = path_to_cropped_dataset + celebrity_name
            cropped_file_name = celebrity_name + str(count) + ".png"
            cropped_file_path = "./dataset/cropped" + "/" + cropped_file_name
            cv2.imwrite(cropped_file_path, roi_color)
            count=count+1
```

```python
cropped_image_dirs = []
celebrity_file_names_dict = {}
for img_dir in img_dirs:
    count = 1
    celebrity_name = img_dir.split('\\')[-1]
    celebrity_file_names_dict[celebrity_name] = []
    for entry in os.scandir(img_dir):
        roi_color = get_cropped_image_if_2_eyes(entry.path)
        if roi_color is not None:
            cropped_folder = path_to_cropped_dataset + "/"+celebrity_name
            if not os.path.exists(cropped_folder):
                os.makedirs(cropped_folder)
                cropped_image_dirs.append(cropped_folder)
                print("Generating cropped images in folder: ",cropped_folder)
            cropped_file_name = celebrity_name + str(count) + ".png"
            cropped_file_path = cropped_folder + "/" + cropped_file_name
            cv2.imwrite(cropped_file_path, roi_color)
            celebrity_file_names_dict[celebrity_name].append(cropped_file_path)
            count += 1
```
